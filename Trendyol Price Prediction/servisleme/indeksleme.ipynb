```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('laptop_trendyol_data.csv')
```

```python
dataset.head()
```

```python



# Belirli sütunlara işlem yapmak için sütunları seçin
columns_to_process = ["Marka", "Ekran Kartı","Çözünürlük","İşlemci Tipi","İşletim Sistemi"]

# Seçili sütunlarda boşlukları kaldırın ve benzersiz değerleri alın
for column in columns_to_process:
    dataset[column] = dataset[column].astype(str).str.replace(' ', '')  # Boşlukları kaldır

    # NaN'leri temizlemeden önce, NaN değerlerini 'None' ile değiştirelim
    dataset[column] = dataset[column].replace('nan', np.nan)

    # NaN'leri temizleyin ve tekil değerleri alın, sıralayın ve CSV olarak kaydedin
    unique_values = dataset[column].dropna().unique()
    unique_values.sort()
    df_unique = pd.DataFrame()
    df_unique[column] = list(unique_values)
    df_unique.to_csv(f"{column}.csv", index=False)
```
